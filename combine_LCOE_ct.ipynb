- Low scenario: 25% capacity ratio → 100 MW renewables + 25 MW battery

- Mid scenario: 50% capacity ratio → 100 MW renewables + 50 MW battery

- High scenario: 100% capacity ratio → 100 MW renewables + 100 MW battery

When thinking about Climate Trace data:

- Low (25%): Best for facilities that can "interrupt" their work (e.g., a pumping station that only runs when the sun is out).

- Mid (50%): A standard hybrid project.

- High (100%): Best for "Baseload" industrial facilities (like a Cement plant or Steel mill) that need power 24/7. These facilities cannot simply turn off when a cloud passes by, so they require the "High" scenario (100% battery ratio) to replace a gas or coal plant.
